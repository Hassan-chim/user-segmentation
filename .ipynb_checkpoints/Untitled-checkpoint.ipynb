## Component 1: Basic Feature Calculators

```python
def calculate_basic_features(user_data):
    features = {}
    features['num_unique_apps'] = user_data['app_name'].nunique()
    features['num_sessions'] = user_data['session_id'].nunique()
    return features
```

```python
def calculate_session_durations(user_data):
    features = {}
    sessions = []
    for session_id, session_data in user_data.groupby('session_id'):
        if len(session_data) > 1:
            session_start = session_data['timestamp'].min()
            session_end = session_data['timestamp'].max()
            session_duration = (session_end - session_start).total_seconds() / 60
            sessions.append(session_duration)
    
    features['avg_session_duration'] = np.mean(sessions) if sessions else 0
    features['total_usage_minutes'] = np.sum(sessions) if sessions else 0
    return features
```

## Component 2: Time of Day Analysis

```python
def get_time_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'
```

```python
def calculate_time_patterns(user_data):
    features = {}
    user_data['hour'] = user_data['timestamp'].dt.hour
    user_data['time_period'] = user_data['hour'].apply(get_time_period)
    time_period_counts = user_data['time_period'].value_counts()
    
    if not time_period_counts.empty:
        features['most_frequent_time'] = time_period_counts.index[0]
        features['morning_ratio'] = (user_data['time_period'] == 'Morning').mean()
        features['afternoon_ratio'] = (user_data['time_period'] == 'Afternoon').mean()
        features['evening_ratio'] = (user_data['time_period'] == 'Evening').mean()
        features['night_ratio'] = (user_data['time_period'] == 'Night').mean()
    else:
        features['most_frequent_time'] = 'Unknown'
        features['morning_ratio'] = 0
        features['afternoon_ratio'] = 0
        features['evening_ratio'] = 0
        features['night_ratio'] = 0
    return features
```

## Component 3: App Category Analysis

```python
def categorize_app(app_name):
    app_name_lower = app_name.lower()
    if 'google' in app_name_lower or 'chrome' in app_name_lower or 'browser' in app_name_lower:
        return 'Browser/Search'
    elif 'facebook' in app_name_lower or 'twitter' in app_name_lower or 'instagram' in app_name_lower:
        return 'Social Media'
    elif 'gmail' in app_name_lower or 'mail' in app_name_lower or 'outlook' in app_name_lower:
        return 'Email'
    elif 'youtube' in app_name_lower or 'netflix' in app_name_lower or 'spotify' in app_name_lower:
        return 'Entertainment'
    elif 'maps' in app_name_lower or 'uber' in app_name_lower:
        return 'Navigation/Transport'
    elif 'calculator' in app_name_lower or 'notes' in app_name_lower or 'calendar' in app_name_lower:
        return 'Productivity'
    else:
        return 'Other'
```

```python
def calculate_app_category_features(user_data):
    features = {}
    user_data['app_category'] = user_data['app_name'].apply(categorize_app)
    app_category_counts = user_data['app_category'].value_counts()
    
    if not app_category_counts.empty:
        features['most_frequent_app_category'] = app_category_counts.index[0]
        categories = ['Browser/Search', 'Social Media', 'Email', 'Entertainment', 
                     'Navigation/Transport', 'Productivity', 'Other']
        for category in categories:
            features[f'{category.lower().replace("/", "_").replace(" ", "_")}_ratio'] = \
                (user_data['app_category'] == category).mean()
    else:
        features['most_frequent_app_category'] = 'Unknown'
    return features
```

## Component 4: Additional Metrics

```python
def calculate_engagement_metrics(user_data, basic_features):
    features = {}
    features['total_events'] = len(user_data)
    num_sessions = basic_features.get('num_sessions', 1)
    features['events_per_session'] = features['total_events'] / num_sessions if num_sessions > 0 else 0
    return features
```

```python
def calculate_app_diversity(user_data):
    features = {}
    app_counts = user_data['app_name'].value_counts()
    if len(app_counts) > 0:
        probabilities = app_counts / app_counts.sum()
        features['app_diversity'] = -np.sum(probabilities * np.log(probabilities + 1e-10))
    else:
        features['app_diversity'] = 0
    return features
```

```python
def calculate_switching_frequency(user_data, session_features):
    features = {}
    total_events = len(user_data)
    total_usage_minutes = session_features.get('total_usage_minutes', 1)
    features['events_per_minute'] = total_events / total_usage_minutes if total_usage_minutes > 0 else 0
    return features
```

# STEP 5: CLUSTERING ANALYSIS 

DETERMINING OPTIMAL NUMBER OF CLUSTERS

```python
print(f"\nDataset size: {X_scaled_df.shape[0]} users")
```

```python
print(f"Testing different numbers of clusters (k)...")

# Determine range of k to test based on dataset size
n_users = X_scaled_df.shape[0]
if n_users < 10:
    max_k = min(5, n_users - 1)
    print(f"Small dataset ({n_users} users), testing k from 2 to {max_k}")
    k_range = range(2, max_k + 1)
elif n_users < 50:
    max_k = min(6, n_users // 2)
    print(f"Medium dataset ({n_users} users), testing k from 2 to {max_k}")
    k_range = range(2, max_k + 1)
else:
    print(f"Large dataset ({n_users} users), testing k from 2 to 6")
    k_range = range(2, 7)

silhouette_scores = []
wcss = []
calinski_scores = []
davies_scores = []

for k in k_range:
    print(f"\nTesting k = {k}:")
    
    # Apply KMeans
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled_df)
    labels = kmeans.labels_
    
    # Calculate metrics (with error handling for small clusters)
    try:
        if len(np.unique(labels)) > 1:
            silhouette_avg = silhouette_score(X_scaled_df, labels)
            calinski = calinski_harabasz_score(X_scaled_df, labels)
            davies = davies_bouldin_score(X_scaled_df, labels)
        else:
            silhouette_avg = -1
            calinski = 0
            davies = float('inf')
    except Exception as e:
        print(f"  Warning: Could not calculate metrics for k={k}: {e}")
        silhouette_avg = -1
        calinski = 0
        davies = float('inf')
    
    silhouette_scores.append(silhouette_avg)
    wcss.append(kmeans.inertia_)
    calinski_scores.append(calinski)
    davies_scores.append(davies)
    
    print(f"  • Silhouette Score: {silhouette_avg:.4f}")
    print(f"  • WCSS (Inertia): {kmeans.inertia_:.2f}")
    print(f"  • Calinski-Harabasz: {calinski:.2f}")
    print(f"  • Davies-Bouldin: {davies:.4f}")

# Find optimal k (skip invalid silhouette scores)
valid_silhouettes = [(k, score) for k, score in zip(k_range, silhouette_scores) if score >= 0]
if valid_silhouettes:
    optimal_k, best_score = max(valid_silhouettes, key=lambda x: x[1])
    print(f"\n✅ Optimal number of clusters: k = {optimal_k}")
    print(f"   (Silhouette score: {best_score:.4f})")
else:
    # Fallback: use elbow method or fixed k
    print("\n⚠️ Could not determine optimal k using silhouette scores")
    print("Using k = 3 as default")
    optimal_k = 3
```

# EVALUATE CLUSTERING RESULTS

```python
print("\n" + "="*70)
print("STEP 7: EVALUATING CLUSTERING RESULTS")
print("="*70)

def evaluate_clustering(X, labels, algorithm_name):
    """Evaluate clustering quality using multiple metrics"""
    unique_labels = np.unique(labels)
    n_clusters = len(unique_labels)
    
    print(f"\n{algorithm_name} Evaluation:")
    print(f"  • Number of clusters: {n_clusters}")
    
    # Check if we have valid clusters (not all noise or single cluster)
    if n_clusters > 1 and (labels != -1).any():
        try:
            # For DBSCAN, filter out noise points for evaluation
            if -1 in unique_labels:
                valid_mask = labels != -1
                X_valid = X[valid_mask]
                labels_valid = labels[valid_mask]
                
                if len(np.unique(labels_valid)) > 1:
                    silhouette = silhouette_score(X_valid, labels_valid)
                    calinski = calinski_harabasz_score(X_valid, labels_valid)
                    davies = davies_bouldin_score(X_valid, labels_valid)
                else:
                    silhouette = -1
                    calinski = 0
                    davies = float('inf')
            else:
                silhouette = silhouette_score(X, labels)
                calinski = calinski_harabasz_score(X, labels)
                davies = davies_bouldin_score(X, labels)
            
            print(f"  • Silhouette Score: {silhouette:.4f} (Higher is better)")
            print(f"  • Calinski-Harabasz: {calinski:.2f} (Higher is better)")
            print(f"  • Davies-Bouldin: {davies:.4f} (Lower is better)")
            
            return silhouette, calinski, davies
            
        except Exception as e:
            print(f"  • Could not calculate metrics: {e}")
            return None, None, None
    else:
        print(f"  • Only {n_clusters} cluster(s), cannot evaluate properly")
        return None, None, None
```

```python
print("\nCLUSTERING ALGORITHM EVALUATION")

# Evaluate each clustering
metrics = {}
for algo, labels in [('KMeans', kmeans_labels), 
                     ('DBSCAN', dbscan_labels), 
                     ('Agglomerative', agg_labels)]:
    metrics[algo] = evaluate_clustering(X_scaled_df, labels, algo)
```

# COMPARE CLUSTERING ALGORITHMS 

```python
print("\n" + "="*70)
print("STEP 10: COMPARING CLUSTERING ALGORITHMS")
print("="*70)

# Create comparison table
print("\n📊 CLUSTERING ALGORITHMS COMPARISON:")
print("-"*60)

# First, let's check what columns we actually have
print("Available cluster columns in user_features_df:")
for col in user_features_df.columns:
    if 'cluster' in col.lower():
        print(f"  • {col}")

# Define mapping between algorithm names and column names
algorithm_mapping = {
    'KMeans': 'kmeans_cluster',
    'DBSCAN': 'dbscan_cluster',
    'Agglomerative': 'agg_cluster'
}

# Evaluate each clustering algorithm
metrics = {}
for algo_name, col_name in algorithm_mapping.items():
    if col_name in user_features_df.columns:
        labels = user_features_df[col_name].values
        unique_labels = np.unique(labels)
        n_clusters = len([l for l in unique_labels if l != -1])  # Exclude noise
        
        print(f"\n{algo_name} ({col_name}):")
        print(f"  • Number of clusters: {n_clusters}")
        
        # Count noise points (only for DBSCAN)
        if algo_name == 'DBSCAN':
            noise_points = (labels == -1).sum()
            print(f"  • Noise points: {noise_points}")
        
        # Check if we have valid clusters for evaluation
        valid_labels = labels[labels != -1] if -1 in unique_labels else labels
        unique_valid = np.unique(valid_labels)
        
        if len(unique_valid) > 1 and len(valid_labels) > 1:
            try:
                # Get corresponding data points (filter out noise for DBSCAN)
                if -1 in unique_labels:
                    valid_mask = labels != -1
                    X_valid = X_scaled_df[valid_mask]
                    labels_valid = labels[valid_mask]
                else:
                    X_valid = X_scaled_df
                    labels_valid = labels
                
                silhouette = silhouette_score(X_valid, labels_valid)
                calinski = calinski_harabasz_score(X_valid, labels_valid)
                davies = davies_bouldin_score(X_valid, labels_valid)
                
                print(f"  • Silhouette Score: {silhouette:.4f}")
                print(f"  • Calinski-Harabasz: {calinski:.2f}")
                print(f"  • Davies-Bouldin: {davies:.4f}")
                
                metrics[algo_name] = {
                    'silhouette': silhouette,
                    'calinski': calinski,
                    'davies': davies,
                    'n_clusters': n_clusters,
                    'noise_points': noise_points if algo_name == 'DBSCAN' else 0
                }
                
            except Exception as e:
                print(f"  • Could not calculate metrics: {e}")
                metrics[algo_name] = None
        else:
            print(f"  • Not enough valid clusters for evaluation")
            metrics[algo_name] = None
    else:
        print(f"\n{algo_name}: Column '{col_name}' not found in DataFrame")
        metrics[algo_name] = None

# Create comparison DataFrame
print("\n" + "="*60)
print("COMPARISON TABLE:")
print("="*60)

comparison_data = []
for algo_name, metric_dict in metrics.items():
    if metric_dict is not None:
        comparison_data.append({
            'Algorithm': algo_name,
            'Silhouette Score': f"{metric_dict['silhouette']:.4f}",
            'Calinski-Harabasz': f"{metric_dict['calinski']:.2f}",
            'Davies-Bouldin': f"{metric_dict['davies']:.4f}",
            'Clusters': metric_dict['n_clusters'],
            'Noise Points': metric_dict.get('noise_points', 0)
        })

if comparison_data:
    comparison_df = pd.DataFrame(comparison_data)
    print(comparison_df.to_string(index=False))
    
    # Determine best algorithm based on silhouette score
    valid_rows = [row for row in comparison_data if row['Silhouette Score'] != 'N/A']
    if valid_rows:
        best_row = max(valid_rows, key=lambda x: float(x['Silhouette Score']))
        print(f"\n✅ BEST PERFORMING ALGORITHM: {best_row['Algorithm']}")
        print(f"   (Highest Silhouette Score: {best_row['Silhouette Score']})")
        
        # Interpret silhouette score
        silhouette_value = float(best_row['Silhouette Score'])
        print(f"\n📊 Silhouette Score Interpretation:")
        print(f"   Score: {silhouette_value:.4f}")
        if silhouette_value > 0.7:
            print("   • Excellent clustering structure")
        elif silhouette_value > 0.5:
            print("   • Reasonable clustering structure")
        elif silhouette_value > 0.25:
            print("   • Weak clustering structure")
        else:
            print("   • No substantial clustering structure")
            print("   • Consider: Fewer clusters, different features, or algorithm")
else:
    print("⚠️ No valid clustering evaluations available")
```
